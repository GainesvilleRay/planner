```python
# libraries
#from collections import defaultdict
import csv
import datetime
from datetime import date
import glob
#from itertools import chain
import os.path, time
```

```python
def main():
    filename = 'submissions.csv'
    pn = []
    with open(filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        headers = next(csv_reader)
        
        for row in csv_reader:
            campname = row[2]
            description = row[4]
            dates = row[5]
            times = row[6]
            cost = row[7]
            website = row[8]
            phone = row[9]
            email = row[10]
            f = open('planner.txt', 'a')
            f.write(campname + ':' + '\n' + description + '\n')
            f.write(cost + '\n')
            f.write(email + '\n\n')
        f.close()
            
if __name__ == '__main__':
    main()
    
f = open("planner.txt", "r")
contents = f.read()
print(contents)
```
