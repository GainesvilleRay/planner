```python
#python3.7
#utf8

"""This script takes a csv file with information related to summer camps
and outputs a text file that is formatted for publication. It was created
Feb. 20, 2019 by Douglas Ray for The Gainesville Sun."""

# libraries
import csv

filename = 'submissions.csv'
```

```python
# Submissions in the category for Art
def cat_art():
    with open(filename) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        headers = next(csv_reader)
        f=open('planner.txt', 'a')
        f.write('ART' + '\n' + '\n')
        
        for row in csv_reader:
            campname = row[2].strip()
            category = row[3].strip()
            description = row[4].strip()
            dates = row[5].strip().replace(' - ', '-')
            times = row[6].strip().replace('P.M.', ' p.m.').replace('A.M.', ' a.m.').replace('  ', ' ')
            location = row[7].strip().replace(', FL', '')
            cost = row[8].strip().replace('.00', '')
            website = row[9].strip().replace('Http://', 'http://')
            phone = row[10].strip()
            email = row[11].strip()
            deadline = row[12].strip()
            
            if "Art" in category:
                f.write(campname + ': ' + description + '\n')
                f.write('Dates: ' + dates + '\n')
                f.write('Times: ' + times + '\n')
                f.write('Location: ' + location + '\n')
                f.write('Cost: ' + cost + '\n')
                f.write('Contact: ' + phone + ', ' + email + '\n')
                f.write('Website: ' + website + '\n')
                f.write('Deadline: ' + '\n\n')
        f.close()
            
```
